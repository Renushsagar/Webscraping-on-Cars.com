```python
#Importing Libraries
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
# Store the URL of the website
website = 'https://www.cars.com/shopping/results/?stock_type=cpo&makes%5B%5D=mercedes_benz&models%5B%5D=&list_price_max=&maximum_distance=all&zip='
```

```python
#Get Request
response = requests.get(website)
```

```python
#Check Status Code
response.status_code
```

```python
# Get Soup Object
soup = BeautifulSoup(response.content, 'html.parser')
soup
```

```python
# Finding all the car details & stored in results
results = soup.find_all('div', class_ = 'vehicle-card') 
```

```python
len(results)
```

```python
#Creating list and appending all the values in the list
cars = []

for car in results:
    Car_name = car.find("h2")
    Car_name = Car_name.get_text() if Car_name is not None else None
    
    mileage_elem = car.find("div", class_="mileage")
    mileage = mileage_elem.get_text() if mileage_elem is not None else None
    
    price_elem = car.find("span", class_="primary-price")
    price = price_elem.get_text() if price_elem is not None else None
    
    rating_elem = car.find('span', class_='sds-rating__count')
    rating = rating_elem.get_text() if rating_elem is not None else None
    
    review_count_elem = car.find("span", class_="test1 sds-rating__link sds-button-link")
    review_count = review_count_elem.get_text() if review_count_elem is not None else None
    
    Dealer_name_elem = car.find("div", class_="dealer-name")
    Dealer_name = Dealer_name_elem.get_text() if Dealer_name_elem is not None else None

    cars.append([Car_name, mileage, price, rating, review_count, Dealer_name])
```

```python
# Creating Dataframe and attributes for cars data
df = pd.DataFrame(cars, columns=["Car Name", "Mileage", "Price", "Rating", "Review Count", "Dealer Name"])
df
```

```python
# Save extracted data in an Excel sheet
df.to_excel("webscraping_Certified_mercedes-benz_cars.xlsx")
```
